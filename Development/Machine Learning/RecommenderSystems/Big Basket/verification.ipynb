# Comparing the results

```python
import pandas as pd

predictions=pd.read_csv('next_orderdate_prediction.csv')
actual=pd.read_excel('BigBasket_test.xlsx')
```

```python
last_dates=[list(set(actual[actual['Member']==value]['lastdate']))[0]  for value in actual['Member'].unique()]
```

```python
predictions['Actual']=last_dates
predictions.drop(predictions.columns[0],axis=1,inplace=True)
```

```python
predictions['Predicted Next Order']=pd.to_datetime(predictions['Predicted Next Order'])
```

```python
predictions['Difference']=predictions['Actual']-predictions['Predicted Next Order']
```

```python
difference=[value.days*value.days for value in predictions['Difference']]
```

```python
predictions['Difference']=difference
```

```python
import math
math.sqrt(predictions['Difference'].sum()/len(difference))
```

```python
predictions.to_csv('Results.csv')
```
