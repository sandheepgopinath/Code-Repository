# Changing the data into required format. 

```python
pCode=neighbourhood_df.PostCode.unique()
```

```python
neighbourhood=pd.DataFrame(columns=neighbourhood_df.columns)
for i in zip(pCode):
    count=neighbourhood_df[neighbourhood_df['PostCode']==str(i[0])].count()[1]
    df=pd.DataFrame(neighbourhood_df[neighbourhood_df['PostCode']==i[0]])
    borough=df['Borough'].unique()[0]
    for j in range(count+1):
        df=pd.DataFrame(neighbourhood_df[neighbourhood_df['PostCode']==i[0]])
        n_append=''
        l=df['Neighbourhood'].values
        for k in range(len(l)):
            n_append=n_append+str(l[k])+','
        n_append=n_append[0:len(n_append)-1]
    temp=pd.DataFrame([[i[0],borough,n_append]],columns=neighbourhood_df.columns)
    neighbourhood=neighbourhood.append(temp)
```

```python
neighbourhood.reset_index(inplace=True)
neighbourhood.drop('index',axis=1,inplace=True)
```
